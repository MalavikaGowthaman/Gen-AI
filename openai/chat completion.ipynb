# ChatCompletion API and Completion API

- Completion API: Hooks you up with text completions from a single prompt. 

- Chat Completion API: Nails it in the chat game, keeping the conversational flow intact

# Completion API

```python
prompt = "What is Python"
response = openai.Completion.create(
    model = "babbage-002",
    prompt = prompt
)
print(response)
```

```python
response = openai.Completion.create(
    model = "gpt-3.5-turbo-instruct",
    prompt = "What is python?"
)
response
```

# ChatCompletion API

```python
prompt = "Hello, How are you?"

response = openai.ChatCompletion.create(
    model = "gpt-4o",
    messages = [
        {"role":"system","content":"You are career advisor"},
        {"role":"user","content":prompt}
    ]
)
print(response)
```

```python
prompt = "Hello, How are you?"

response = openai.ChatCompletion.create(
    model = "gpt-4o",
    messages = [
        {"role":"system","content":"You are career advisor"},
        {"role":"user","content":prompt}
    ]
)
print(response["choices"])
```

```python
print(response["choices"][0])
```

```python
print(response["choices"][0]["message"])
```

```python
print(response["choices"][0]["message"]["content"])
```
